Displaying the Image,
and a few basic functions


```python
#'plt.imshow' is used to display an image or a 2D array

plt.imshow(img)
```

NOTE: Matplotlib and OpenCV expect different orders for color channels in images:

OpenCV (BGR Order): OpenCV, by default, represents color images using the BGR (Blue, Green, Red) color channel order. In other words, when you load an image using cv2.imread, it arranges the color channels in the order of BGR.

Matplotlib (RGB Order): Matplotlib, on the other hand, expects images to have the RGB (Red, Green, Blue) color channel order. When you display an image using plt.imshow or other Matplotlib functions, it assumes that the color channels are arranged in the order of RGB.

```python
# 'cv2.cvtColor()' is a function used to convert the color spaces of images to different forms
# First argument is the source image
# Second argument is the color transformation, in this case BGR to RGB

fixedImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fixedImg)
```

```python
#Reading images in Grayscale

imgGray = cv2.imread('Images\hills.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(imgGray, cmap='gray')

# When you specify cmap='gray', you are telling Matplotlib to use this grayscale colormap for displaying your image.
# If not, Matplotlib uses the default colormap, which is typically a continuous gradient of colors.
```

RESIZING THE IMAGE

```python
resizedImg = cv2.resize(fixedImg, (500,500))

#(500,500) indicate the width and height respectively.

plt.imshow(resizedImg)
```

 FLIPPING THE IMAGE

```python
flipImg = cv2.flip(fixedImg, 0)
plt.imshow(flipImg)


#NOTE: the second argument '0' implies the flip along horizontal middle axis.
#                          '1' for a flip along vertical middle axis.
#                          '-1' for a flip along both the axes.
```

SAVE THE MODIFIED NUMPY ARRAY TO AN IMAGE FILE


```python
cv2.imwrite('Images\saved_img.jpeg', fixedImg)
```
