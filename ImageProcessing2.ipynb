```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#A function to load the required blank image to perform operations
def load_img():
    blank_image = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_image,text="ABCDE", org=(50,300),fontScale=5, color=(255,255,255),fontFace=font,thickness=25)
    return blank_image

#A function to display the image
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = load_img()
display_img(img)
```

## Morphological Operators
Morphological Operators are a set of Kernels that can achieve a variety of effects, such as reducing noise.

Certain Operators are very good at reducing Black points on a white background or vice versa.

Few can also achieve erosion or dilation effect that can add or erode from an existing image.

```python
#1) Erode:
#   

kernel = np.ones((5,5), dtype=np.uint8)

eroded = cv2.erode(img,kernel,iterations=4) #Higher iterations == Higher erosion
display_img(eroded)
```

```python
#Adding some white noise to the original image
img = load_img()
white_noise = np.random.randint(low=0,high=2,size=(600,600))
white_noise = white_noise * 255
display_img(white_noise)
```

```python
mixed_img = white_noise + img
display_img(mixed_img)
```

```python
opening = cv2.morphologyEx(mixed_img, cv2.MORPH_OPEN, kernel)
display_img(opening)
#Mostly used for removing background noise
```

```python
#Creating a foreground noise - black_noise:
img = load_img()
black_noise = np.random.randint(low=0,high=2,size=(600,600))
black_noise = black_noise * -255
black_noised_img = img + black_noise
black_noised_img[black_noised_img== -255] =0 #0 should be the smallest value in a black noised image
display_img(black_noised_img)
```

```python
closing = cv2.morphologyEx(black_noised_img,cv2.MORPH_CLOSE, kernel)
display_img(closing)
```
